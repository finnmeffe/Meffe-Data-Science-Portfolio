# Data loading and exploration
Load California Housing dataset

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

housing = fetch_california_housing()
```

Create dataframe for features and target variable

```python
X = pd.DataFrame(housing.data, columns=housing.feature_names) 
print(X)
y = pd.Series(housing.target, name='med_house_value')
```

## Initial exploration of the dataset: 
First, display the first 5 rows

```python
# display the first 5 rows
print("Dataset preview: ")
print(X.head())
```

Display the names of features and check for any NA values

```python
# give names of features; check for NA
print("\n Feature names: ")
print(list(X.columns))
print("\n Check for NA values by variable: ")
print(X.isna().sum())
```

Show some basic summary statistics:

```python
# summary stats
print("\n Basic summary statistics: ")
print(X.describe())
```

# Linear Regression on Unscaled Data

Set up the linear regression with testing and training data and make first predictions

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# split 80-20 train test
X_train_raw, X_test_raw, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
# can use random_state=int to control seed

# set up linear regression
lin_reg_raw = LinearRegression()
lin_reg_raw.fit(X_train_raw, y_train)

# predict from training
y_pred_raw = lin_reg_raw.predict(X_test_raw)
```

Evaluate the model performance with $MSE$, $RMSE$, and $R^2$

```python
from sklearn.metrics import mean_squared_error, root_mean_squared_error, r2_score

mse_raw = mean_squared_error(y_test, y_pred_raw)
rmse_raw = root_mean_squared_error(y_test, y_pred_raw)
r2_raw = r2_score(y_test, y_pred_raw)

print("Metrics for model performance: ")
print(f"MSE: {mse_raw:.4f}")
print(f"RMSE: {rmse_raw:.4f}")
print(f"R² Score: {r2_raw:.4f}")

print("\n Model coefficients: ")
print(pd.Series(lin_reg_raw.coef_,
                index=X.columns))
```

### What does the R² score tell us about model performance?

The calculated R² score is 0.6122, which indicates that the model is able to explain over half of the variance - a reasonably good fit.

### Which features seem to have the strongest impact on predictions based on the model’s coefficients?

Going solely off of the value of estimated coefficients, it seems as though the average number of bedrooms and median income have the largest effect on the house value. Intuitively, this makes sense, as median income likely reflects strong neighborhood amenities, and the number of bedrooms is increasing in house size.

### How well do the predicted values match the actual values?

The root mean squared error for the model is 0.7355, which gives an idea of the degree to which the predicted values deviate from the actual values in units of y. Given that the mean y value is approximately 2 with standard deviation of 1.15, the predicted values do not seem to match the actual values extremely well.


# Feature Selection and Simplified Model

I select median income, average rooms, and average bedrooms to be part of the simplified model, as these have the largest coefficients by magnitude. Without knowing anything else about the coefficients (e.g. SE or t-stat) it is impossible to know how precise these coefficients are, so I use a simple selection based on size.

```python
# create subset of X data and repeat the same analysis as above
X_sub = X[['MedInc','AveRooms','AveBedrms']]

X_train_sub, X_test_sub, y_train_sub, y_test_sub = train_test_split(X_sub, y, test_size = 0.2, random_state = 20)

lin_reg_sub = LinearRegression()
lin_reg_sub.fit(X_train_sub, y_train_sub)
y_pred_sub = lin_reg_sub.predict(X_test_sub)

mse_sub = mean_squared_error(y_test_sub, y_pred_sub)
rmse_sub = root_mean_squared_error(y_test_sub, y_pred_sub)
r2_sub = r2_score(y_test_sub, y_pred_sub)

print("Metrics for simplified model performance: ")
print(f"MSE: {mse_sub:.4f}")
print(f"RMSE: {rmse_sub:.4f}")
print(f"R² Score: {r2_sub:.4f}")

print("\n Model coefficients: ")
print(pd.Series(lin_reg_sub.coef_,
                index=X_sub.columns))
```

One can directly compare the 3 calculated performance metrics between models: the simplified model has both a higher MSE and RMSE, as well as a lower R². This the fitted values are, on average, further away from the real values of y (house value), and the model is able to capture less of the variance in y. In practice, it may be easier to use this model, as collecting data for 3 variables is likely far easier than the full set of 8: this may be particularly relevant if the data source only contains some subset of variables (e.g. a realty data containing data on rooms but not the population of the area).
